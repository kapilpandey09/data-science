```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

```python
# Scatter plot: Used to visualize the relationship between two variables, with each data point represented by a point on the plot.
import seaborn as sns
iris = sns.load_dataset('iris')
sns.scatterplot(x = iris.sepal_length,y = iris.sepal_width)
```

```python
# Line plot: Used to visualize the relationship between two or more variables over time.
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
sns.lineplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
# He displot() function in Seaborn is a high-level function that provides a convenient way to visualize the distribution of a univariate or bivariate dataset. It can be used to create a variety of plot types, including histograms, kernel density estimates (KDEs), and rug plots.
import seaborn as sns
iris = sns.load_dataset("iris")
sns.displot(x = iris['sepal_length'])
```

```python
# pairplot: It can be used to visualize the relationships between all pairs of variables in a dataset, and to identify patterns, trends, and outliers.
import seaborn as sns

# Load the data
data = sns.load_dataset("tips")

# Create a pairplot of all variables in the tips dataset
sns.pairplot(data=data)


```

```python
# jointplot: A joint plot is a combination of a scatter plot and a density plot, and it can be used to visualize the relationship between two variables and the distribution of each variable.
import seaborn as sns
tips = sns.load_dataset("tips")
tips
sns.jointplot(x = "total_bill", y = 'tip', data = tips)
```

```python
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

Note: timepoint, signal, event, and region are columns in the fmri dataset.
```

```python
import seaborn as sns

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Plot a line plot using x = "timepoint" and y = "signal" for different events and regions
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
```

```python
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.
```

```python
import seaborn as sns
titanic = sns.load_dataset("titanic")
sns.boxplot(x = "pclass", y = "age", showmeans=True, data=titanic)
sns.boxplot(x = "pclass", y = "fare", showmeans=True, data=titanic)
# import matplotlib.pyplot as plt
# plt.title("Box-Plot")
# plt.show()
```

```python
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
```

```python
import seaborn as sns
diamond = sns.load_dataset("diamonds")
sns.histplot(x= "price", hue="cut", data= diamond, kde= "True" )
```

```python
Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset("iris")
sns.pairplot(iris, hue="species")
plt.show()
```

```python
Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot("month", "year", "passengers")
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d", linewidths=0.5)
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Flights Dataset Heatmap")

# Display the plot
plt.show()
```
