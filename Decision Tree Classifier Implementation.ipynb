# Decision Tree Classifier Implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
print(dataset.DESCR)
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
dataset.target
```

```python
# Independent and Dependent features
X = df.iloc[:,:-1]
Y = dataset.target
```

```python
X,Y
```

```python
## train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier()
decision_tree
```

```python
decision_tree.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(decision_tree,filled=True)
```

```python
# post_puning with max_depth = 2
post_prunning = DecisionTreeClassifier(max_depth=2)
post_prunning.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(16,10))

tree.plot_tree(post_prunning,filled=True)
```

```python
## prediction
y_pred = post_prunning.predict(X_test)
```

```python
y_pred,y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score, classification_report
```

```python
score = accuracy_score(y_test,y_pred )
print(score)
print(classification_report(y_pred,y_test))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
y_pred_dec = decision_tree.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred_dec))
print(accuracy_score(y_test, y_pred_dec))
print(classification_report(y_test, y_pred_dec))
```

# DecisionTree Preruning And Hyperparameter Tunning for huge data

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
parameter={
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treeclassifier=DecisionTreeClassifier()
clf = GridSearchCV(treeclassifier,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
y_pred = clf.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```
