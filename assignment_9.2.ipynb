Matplotlib is a Python library for data visualization. 
It is used to create a wide variety of plots, including line plots, bar plots, scatter plots, pie charts, histograms, and more. Matplotlib is a powerful tool for data exploration and communication, and it is widely used by scientists, engineers, and data scientists.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:

Line plot: A line plot is used to show the relationship between two variables, such as time and temperature.

Bar plot: A bar plot is used to compare the values of different categories.

Scatter plot: A scatter plot is used to show the relationship between two variables, where each data point is represented by a point on the plot.

Pie chart: A pie chart is used to show the relative proportions of different categories.

Histogram: A histogram is used to show the distribution of a continuous variable.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Line plot
x = [1,2,3,4,5]
y = [2,3,4,5,6]
plt.plot(x,y)
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("This is a plot")
plt.show()
```

```python
# Bar Plot : 
import matplotlib.pyplot as plt
import numpy as np
x = ['a', 'b', 'c', 'd', 'e']
y = [1,2,6,7,9]
plt.bar(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bar-Plot")
```

```python
# Scatter plot
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(54)
y = np.random.rand(54)
color = np.random.rand(54)
plt.scatter(x,y, c= color, s = 340, alpha = .6, marker='v')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter plot")
plt.show()
```

```python
# Pie Plot
import matplotlib.pyplot as plt
import numpy as np
slices = ["Apples", "Oranges", "Bananas"]
values = [10, 15, 20]

plt.pie(values, labels=slices, autopct="%1.1f%%")
plt.title("Pie Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = [1,2,3,3,4,1,4,3,5,8,9,6,7,0,2,1,4,1,2,3,3,4,5,6,6,6]

plt.hist(x)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Histogram-Plot")
plt.show()
```

```python
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.
```


A scatter plot is a type of plot that shows the relationship between two variables. It is created by plotting each data point as a point on a two-dimensional graph, with one variable on the horizontal axis and the other variable on the vertical axis.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("This is scatter plot")
plt.show()
```

```python
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

The subplot() function is used to create multiple plots on a single figure. It is a convenient way to create complex figures with multiple plots, such as scatter plots, line plots, and bar plots.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)  # Create a 2x2 grid of subplots, and this is the first subplot
plt.plot(x, y1)
plt.title('Line 1')

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)  # This is the second subplot
plt.plot(x, y2)
plt.title('Line 2')

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)  # This is the third subplot
plt.plot(x, y3)
plt.title('Line 3')

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)  # This is the fourth subplot
plt.plot(x, y4)
plt.title('Line 4')

plt.tight_layout()  # To improve subplot spacing
plt.show()
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

A box plot, also known as a box and whisker plot, is a graphical method for displaying the distribution of data. It shows the five-number summary of a set of data, which includes the minimum, the first quartile, the median, the third quartile, and the maximum.

```python
import matplotlib.pyplot as plt
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
labels = ['Box 1', 'Box 2']

plt.boxplot(data, labels=labels)

plt.title('Box Plot of Box 1 and Box 2')
plt.ylabel('Values')

plt.show()
```
