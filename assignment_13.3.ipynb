# Q1. What is Min-Max scaling, and how is it used in data preprocessing? Provide an example to illustrate its  application. 

Min-Max scaling is a data normalization technique used in data preprocessing. Its purpose is to scale and transform the features of a dataset so that they fall within a specific range typically between 0 and 1




```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
data = np.array([[18.0, 7.0, 8.0],
                 [9.0, 10.0, 11.0],
                 [12.0, 13.0, 14.0]])

min_max = MinMaxScaler()
min_max_data = min_max.fit_transform(data)

print("Original data:", data)
print("\nScaled data:\n", min_max_data)

```

## Q3. What is PCA (Principle Component Analysis), and how is it used in dimensionality reduction? Provide an example to illustrate its application.

Principal Component Analysis (PCA) is a technique used in statistics and machine learning for dimensionality reduction. Its primary goal is to transform high-dimensional data into a lower-dimensional representation while retaining as much of the original variance as possible. This is achieved by identifying the principal components, which are linear combinations of the original features that capture the most significant variability in the data.



```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Generate 2D data
np.random.seed(42)
data = np.random.rand(100, 2) * 10 + np.array([3, 7])

# Apply PCA
pca = PCA(n_components=1)
transformed_data = pca.fit_transform(data)

# Plot original and transformed data
plt.scatter(data[:, 0], data[:, 1], label='Original Data')
plt.scatter(transformed_data, np.zeros_like(transformed_data), label='Transformed Data', alpha=0.7)
plt.legend()
plt.title('PCA Example: Dimensionality Reduction')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()


```

In this example, PCA is used to reduce the 2D data to a one-dimensional representation along the principal component. The transformed data retains much of the original variability along the most significant direction in the data.

PCA (Principal Component Analysis) is closely related to feature extraction. PCA can be effectively used for Feature Extraction by identifying the principal components of a dataset, which represent the directions of maximum variance in the data. These principal components can then be used as new features, capturing the most significant patterns and variations in the data. By reducing the dimensionality of the data using PCA, we can eliminate redundant or irrelevant features, leading to a more efficient and interpretable machine learning model.


Consider the example of image recognition. An image can be represented as a high-dimensional vector, where each pixel value corresponds to a feature. PCA can be applied to this high-dimensional image representation to extract the most significant features, such as edges, shapes, and textures. These extracted features can then be used to train a machine learning model to classify images more accurately.


In summary, PCA and Feature Extraction are complementary techniques that can be used together to improve the performance of machine learning models. PCA provides a powerful method for dimensionality reduction, while Feature Extraction focuses on selecting the most relevant features from the data. By combining PCA with Feature Extraction, we can achieve a more efficient and effective representation of the data, leading to better-performing machine learning models.

## Q7. For a dataset containing the following values: [1, 5, 10, 15, 20], perform Min-Max scaling to transform the values to a range of -1 to 1.

```python
import numpy as np

# Given dataset
data = np.array([1, 5, 10, 15, 20])


min_new, max_new = -1, 1


data_scaled = (data - np.min(data)) / (np.max(data) - np.min(data)) * (max_new - min_new) + min_new

print("Original Data:", data)
print("Scaled Data (-1 to 1):", data_scaled)
```

Determining the number of principal components (PCs) to retain in a dataset using Principal Component Analysis (PCA) depends on the specific dataset and the desired level of dimensionality reduction. However, there are some general guidelines that can be followed.

Eigenvalue Threshold: One common approach is to retain PCs with eigenvalues greater than a certain threshold. This threshold can be set to 1, which means only PCs that explain more than 100% of the variance in the data are retained. A more common choice is to set the threshold to 0.7 or 0.8, which means only PCs that explain at least 70% or 80% of the variance are retained.

1. Scree Plot: Another method is to use a scree plot, which is a plot of the eigenvalues of the PCs. The scree plot will typically show a sharp elbow, where the eigenvalues start to level off. The number of PCs to the left of the elbow are the ones that should be retained.

2. Cumulative Variance: Finally, you can also look at the cumulative variance of the PCs. The cumulative variance is the sum of the explained variances of the PCs. The number of PCs that explain a certain percentage of the variance (e.g., 90%) are the ones that should be retained.

3. In the case of the dataset containing the following features: [height, weight, age, gender, blood pressure], it is likely that you would want to retain at least 2 or 3 PCs. This is because these features are all related to each other to some extent, and PCA can be used to capture these relationships.

Here is a more specific recommendation for the number of PCs to retain:

##### Retain 3 PCs: This is a good choice if you want to capture the most important relationships between the features.
